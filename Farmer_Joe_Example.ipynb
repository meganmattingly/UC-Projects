```python
#Import libraries
from ortools.linear_solver import pywraplp as glp

#Create model
model = glp.Solver.CreateSolver("GLOP")

#Define variables
x_1 = model.NumVar(0, 999, "x_1")
x_2 = model.NumVar(0, 999, "x_2")

#Define objective
obj = model.Objective()
obj.SetCoefficient(x_1, 100)
obj.SetCoefficient(x_2, 30)
obj.SetMaximization()

#Define constraints
constraint_1 = model.Constraint(30, 999, "c_1")
constraint_1.SetCoefficient(x_2, 10)

constraint_2 = model.Constraint(-999, 7, "c_2")
constraint_2.SetCoefficient(x_1, 1)
constraint_2.SetCoefficient(x_2, 1)

constraint_3 = model.Constraint(-999, 40, "c_3")
constraint_3.SetCoefficient(x_1, 10)
constraint_3.SetCoefficient(x_2, 4)

#Solve the model
status = model.Solve()

if(status == glp.Solver.OPTIMAL):
  print("Total Profit", round(obj.Value(), 2))
  print("x_1 = ", round(x_1.solution_value(), 2))
  print("x_2 = ", round(x_2.solution_value(), 2))
else:
  print("Warning: Model not solved to optimality")
```
