```python
#Load data
numPlants = 3
numDC = 4
P = ["Plant " + str(i + 1) for i in range(0, numPlants)]
DC = ["DC " + str(j + 1) for j in range(0, numDC)]
capacity = [5000, 6000, 2500]
demand = [6000, 4000, 2000, 1500]
cost = [[3, 2, 7, 6],
        [6, 5, 2, 3],
        [2, 5, 4, 5]]
```

```python
#Model
#Import libraries
from ortools.linear_solver import pywraplp as glp

#Create model
model = glp.Solver.CreateSolver("GLOP")

#Define matrices for variables
x = [[model.NumVar(0, 9999, "x " + i + " " +j) for j in DC] for i in P]

#Define objective
obj = model.Objective()
for i in range(0, numPlants):
  for j in range (0, numDC):
    obj.SetCoefficient(x[i][j], cost[i][j])

obj.SetMinimization()

#Define constraints
cap_constraint = [model.Constraint(-9999, capacity[i], "C" + P[i]) for i in range(0, numPlants)]
for i in range(0, numPlants):
  for j in range (0, numDC):
    cap_constraint[i].SetCoefficient(x[i][j], 1)

demand_constraint = [model.Constraint(demand[j], 9999, "D" + DC[j]) for j in range(0, numDC)]
for j in range (0, numDC):
  for i in range(0, numPlants):
    demand_constraint[j].SetCoefficient(x[i][j], 1)

#Solve the model
status = model.Solve()

if(status == glp.Solver.OPTIMAL):
  print("Total Cost", round(obj.Value(), 2))
  for i in range(0, numPlants):
    for j in range (0, numDC):
      if(x[i][j].solution_value() > 0):
        print("Send", round(x[i][j].solution_value(), 2), "items from",
              P[i], "to", DC[j])
else:
  print("Warning: Model not solved to optimality")
```
